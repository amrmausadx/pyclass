## Dictionary

القاموس هو نوع بيانات يُخزن أزواج من المفاتيح والقيم.

```python
person = {'name': 'Ali', 'age': 30}  # قاموس يحتوي على مفتاح name و age
print(person['name'])  # طباعة قيمة المفتاح name

person['age'] = 31  # تعديل قيمة المفتاح age
print(person['age'])  # طباعة قيمة المفتاح age

person['city'] = 'Riyadh'  # إضافة مفتاح جديد city
print(person)  # طباعة القاموس بالكامل

person.pop('age')  # حذف المفتاح age
print(person)  # طباعة القاموس بالكامل

del person['city']  # حذف المفتاح city
print(person)  # طباعة القاموس بالكامل

person.clear()  # إزالة كل العناصر من القاموس
print(person)  # طباعة القاموس بالكامل

person.update({'name': 'Ahmed', 'age': 32})  # تعديل القاموس
print(person)  # طباعة القاموس بالكامل

person.keys()  # طباعة المفاتيح
print(person.keys())  # طباعة المفاتيح
```

## Loop

الحلقة تُستخدم لتكرار تنفيذ كود معين عدة مرات.

```python
for i in range(3):
    print(i)  # ستقوم بطباعة الأرقام من 0 إلى 2
print('range(3)\n')

for i in range(1, 3):
    print(i)  # ستقوم بطباعة الأرقام من 1 إلى 2
print('range(1,3)\n')

for i in range(1, 3, 2):
    print(i)  # ستقوم بطباعة الأرقام من 1 إلى 2 بخطوة 2
print('range(1,3,2)\n')


for i in range(33, 0, -1):
    print(i)  # ستقوم بطباعة الأرقام من 33 إلى 1 بخطوة 1
print('range(33, 0, -1)\n')
```

## Loop - While

حلقة while تستمر في التكرار طالما أن الشرط صحيح.

```python
x = 0
while x < 3:
    print(x)  # طباعة x طالما أنها أقل من 3
    x += 1  # زيادة قيمة x بمقدار 1 في كل مرة
print('while x < 3\n')

x = 0
while x < 30:
    x += 1 # زيادة قيمة x بمقدار 1 في كل مرة
    if x % 2 == 0:continue
    print(x) 
    if x == 15:
        break
```

## Recursion

الاستدعاء الذاتي هو أن تقوم دالة باستدعاء نفسها.

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)  # استدعاء ذاتي

print(factorial(5))
```

```python
print(5*factorial(4))
print(5*4*factorial(3))
print(5*4*3*factorial(2))
print(5*4*3*2*factorial(1))
print(5*4*3*2*1*factorial(0))
print(5*4*3*2*1*1)
```

Write Python code to manage a simple hotel room booking system using a dictionary. Your program should:

-   Read existing hotel names and available rooms from a file called hotels.txt.

-   Add 3 hotels (e.g., Hilton, Marriott, Four Seasons) with a predefined number of available rooms.

-   Display the available rooms for each hotel.

-   Allow the user to book or cancel a room.

-   Save the user's action (booked or cancelled) into a file called ticket.txt, including the hotel name.

-   Update the available rooms in the hotels.txt file accordingly.



```python
# نظام مبسط لحجز فندق بدون تاريخ

# المرحلة 1: تحميل بيانات الفنادق من ملف "hotels.txt"
hotels = {}

try:
    with open("hotels.txt", "r") as file:
        for line in file:
            name, rooms = line.strip().split(",")
            hotels[name.strip()] = int(rooms.strip())
except FileNotFoundError:
    pass  # إذا لم يوجد الملف، نبدأ ببيانات جديدة

# المرحلة 2: إضافة 3 فنادق
hotels["Hilton"] = 5
hotels["Marriott"] = 3
hotels["Four Seasons"] = 7

# المرحلة 3: عرض الغرف المتاحة
print("الغرف المتاحة:\n")
for name, rooms in hotels.items():
    print(f"{name}: {rooms} غرفة متاحة")

# المرحلة 4: حجز أو إلغاء
action = input("\nهل ترغب في الحجز أو الإلغاء؟ (book/cancel): ").strip().lower()
hotel_name = input("أدخل اسم الفندق: ").strip()

if hotel_name not in hotels:
    print("الفندق غير موجود.")
else:
    if action == "book":
        if hotels[hotel_name] > 0:
            hotels[hotel_name] -= 1
            print("تم حجز الغرفة بنجاح.")
            ticket_action = "BOOKED"
        else:
            print("لا توجد غرف متاحة.")
            ticket_action = "FAILED BOOKING"
    elif action == "cancel":
        hotels[hotel_name] += 1
        print("تم إلغاء الحجز بنجاح.")
        ticket_action = "CANCELLED"
    else:
        print("إجراء غير صحيح.")
        ticket_action = "INVALID"

    # المرحلة 5: حفظ سجل الحجز في ملف
    with open("ticket.txt", "a") as ticket_file:
        ticket_file.write(f"{ticket_action}: {hotel_name}\n")

    # المرحلة 6: حفظ عدد الغرف المحدث
    with open("hotels.txt", "w") as hotel_file:
        for name, rooms in hotels.items():
            hotel_file.write(f"{name},{rooms}\n")
```
